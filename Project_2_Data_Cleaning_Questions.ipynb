```python
file_pth = "/content/drive/MyDrive/Car_Sales_Cleaned.csv"
df = pd.read_csv(file_pth)
```

**Q1: Which product line generates the highest total sales revenue?**

```python
df.groupby("PRODUCTLINE")["REVENUE"].sum().sort_values(ascending=False)
```

**Q2: Which geographical territory contributes the most sales?**

```python
df.groupby("TERRITORY")["REVENUE"].sum().sort_values(ascending=False)
```

**Q3: Which country has the highest sales volume?**

```python
df.groupby("COUNTRY")["REVENUE"].sum().sort_values(ascending=False)
```

**Q4: What is the average order value by product line?**

```python
df.groupby("PRODUCTLINE")["REVENUE"].mean().sort_values(ascending=False)
```

**Q5: Are sales increasing or decreasing over time?**

```python
monthly_sales = df.groupby(["YEAR_ID","MONTH_ID"])["REVENUE"].sum()
monthly_sales
```

**Q6: Which customers generate the highest revenue?**

```python
df.groupby("CUSTOMER")["REVENUE"].sum().sort_values(ascending=False).head(10)
```

**Q7: Which year recorded the highest sales?**

```python
df.groupby("YEAR_ID")["REVENUE"].sum()
```

**Q8: Which months generate the highest revenue?**

```python
df.groupby("MONTH_ID")["REVENUE"].sum().sort_values(ascending=False)
```

**Q9: Which quarter contributes the highest sales?**

```python
df.groupby("QTR_ID")["REVENUE"].sum().sort_values(ascending=False)
```

**Q10: Which cities generate the highest revenue?**

```python
df.groupby("CITY")["REVENUE"].sum().sort_values(ascending=False)
```
